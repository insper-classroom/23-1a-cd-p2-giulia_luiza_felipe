```python
data = pd.read_csv('data.csv',encoding='ISO-8859-1', sep=';', low_memory=False)
data
```

```python
filtra_ano = data['NU_ANO']==2006
datacerta = data.loc[filtra_ano,:]
datacerta
```

```python
datacerta.loc[datacerta.TP_DEPENDENCIA_ADM_ESCOLA<4,"TIPO"]="PU"
datacerta.loc[datacerta.TP_DEPENDENCIA_ADM_ESCOLA==4,"TIPO"]="PV"
datacerta.TIPO.value_counts()
```

Quantitava com quantitativa

```python
cor = datacerta['NU_MEDIA_TOT'].corr(datacerta['NU_PARTICIPANTES'])
cor
```

```python
#media com numero de participantes 
plt.scatter(datacerta.NU_MEDIA_TOT,datacerta.NU_PARTICIPANTES)
plt.xlabel("media ENEM")
plt.ylabel("participantes de cada escola")
plt.show()
```

quantitativa com qualitativas

```python
#participantes por escola publica ou privada
faixa=arange(0,100,5)
plt.figure(figsize=(13,4))
plt.subplot(121)
plt.hist(datacerta.NU_MEDIA_TOT[datacerta.TIPO=="PV"],bins=faixa,edgecolor="white",density=True,color="violet")
plt.title('Medias em escolas privadas')
plt.ylabel('densidade')
plt.xlabel('media')



plt.subplot(122)
plt.hist(datacerta.NU_MEDIA_TOT[datacerta.TIPO=="PU"],bins=faixa,edgecolor="white",density=True,color="green")
plt.title('Medias em escolas públicas')
plt.ylabel('densidade')
plt.xlabel('media')
plt.tight_layout()
plt.show()
```
